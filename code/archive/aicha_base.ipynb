```python
def compute_beta_band_power(frequencies1, psd1):
    # Define the frequency range for the beta band
    beta_band_range = (13, 35)

    # Find indices corresponding to frequencies within the beta band range
    beta_band_indices1 = np.where((frequencies1 >= beta_band_range[0]) & (frequencies1 < beta_band_range[1]))[0]

    # Integrate PSD values within the beta band using the trapezoidal rule
    beta_band_power1 = np.trapz(psd1[beta_band_indices1], frequencies1[beta_band_indices1])

    # Return the computed beta band power
    return beta_band_power1

# Example usage:
# Assume frequencies and psd are your PSD values computed using Welch's method

# Compute beta band power
beta_band_power1 = compute_beta_band_power(frequencies1, psd1)
print(f'Beta band power M0S0: {beta_band_power1}')
```

```python
# Recalculate Correlations
def calculate_correlations(df_x, df_y):
    spearman_results = []
    pearson_results = []
    
    for column_x in df_x.columns:
        if column_x in df_y.columns:
            # Spearman correlation
            spearman_corr, spearman_pval = spearmanr(df_x[column_x], df_y[column_x])
            spearman_results.append((column_x, spearman_corr, spearman_pval))
            
            # Pearson correlation
            pearson_corr, pearson_pval = pearsonr(df_x[column_x], df_y[column_x])
            pearson_results.append((column_x, pearson_corr, pearson_pval))
    
    return spearman_results, pearson_results

# Calculate correlations
spearman_beta_ema, pearson_beta_ema = calculate_correlations(df1, df2)
spearman_beta_updrs, pearson_beta_updrs = calculate_correlations(df1, df3)
spearman_ema_updrs, pearson_ema_updrs = calculate_correlations(df2, df3)

# Print results
def print_results(label, spearman_results, pearson_results):
    print(f"\nSpearman Correlation: {label}")
    for cond, corr, pval in spearman_results:
        print(f"{cond}: Spearman rho = {corr:.2f}, p = {pval:.2e}")

    print(f"\nPearson Correlation: {label}")
    for cond, corr, pval in pearson_results:
        print(f"{cond}: Pearson r = {corr:.2f}, p = {pval:.2e}")

print_results("Beta vs EMA", spearman_beta_ema, pearson_beta_ema)
print_results("Beta vs UPDRS", spearman_beta_updrs, pearson_beta_updrs)
print_results("EMA vs UPDRS", spearman_ema_updrs, pearson_ema_updrs)

# Plot function with annotations
def plot_scatter(df_x, df_y, xlabel, ylabel, title, spearman_results):
    markers = ['o', 's', '^', 'D']
    
    plt.figure(figsize=(12, 8))

    for i, column in enumerate(df_x.columns):
        if column in df_y.columns:
            plt.scatter(df_y[column], df_x[column], label=column, marker=markers[i])
        else:
            print(f"Column '{column}' is missing in df_y.")  # This should not happen

    # Customize x-axis tick labels
    ticks = plt.xticks()[0]
    labels = [f'{tick}' for tick in ticks]
    plt.xticks(ticks=ticks, labels=labels)

    plt.axhline(0, color='black', linewidth=1)
    plt.axvline(0, color='black', linewidth=1)
    plt.xlabel(xlabel, fontsize=14)
    plt.ylabel(ylabel, fontsize=14)
    plt.title(title, weight='bold', fontsize=20)
    plt.grid(True)
    plt.legend()

    # Annotate with correlation results
    annotations = '\n'.join([f"{cond}: Spearman rho = {corr:.2f}, p = {pval:.2e}" 
                             for cond, corr, pval in spearman_results])
    
    plt.figtext(0.5, -0.03, annotations, ha='center', va='top', fontsize=12,
                bbox=dict(boxstyle="round", edgecolor='black', facecolor='white'))
    plt.show()

# Plot Beta vs EMA
plot_scatter(df2, df1, 'Normalized EMA Score Difference', 'Normalized Beta Power Difference', 
             'Beta Power Correlated with Normalized EMA Scores', spearman_beta_ema)

# Plot Beta vs UPDRS
plot_scatter(df3, df1, 'Normalized UPDRS subscore Difference', 'Normalized Beta Power Difference', 
             'Beta Power Correlated with Normalized UPDRS Subsccores', spearman_beta_updrs)

# Plot EMA vs UPDRS
plot_scatter(df3, df2, 'Normalized EMA Score Difference', 'Normalized UPDRS subscore Difference', 
             'UPDRS Subsccores Correlated with Normalized EMA Scores', spearman_ema_updrs)
```
